```python
nba_draft = pd.read_csv("2007-2016_nba_draft_rd1-2.csv", engine='python')
nba_draft.head()
```

```python
draft_condensed = nba_draft[["Player", "Year", "Tm", "College"]]
draft_condensed.head()
```

```python
#Can convert to list by adding .values.tolist() at the end of each
first_year = draft_condensed.loc[draft_condensed["Year"] == 2007].copy()
second_year = draft_condensed.loc[draft_condensed["Year"] == 2008].copy()
third_year = draft_condensed.loc[draft_condensed["Year"] == 2009].copy()
fourth_year = draft_condensed.loc[draft_condensed["Year"] == 2010].copy()
fifth_year = draft_condensed.loc[draft_condensed["Year"] == 2011].copy()
sixth_year = draft_condensed.loc[draft_condensed["Year"] == 2012].copy()
seventh_year = draft_condensed.loc[draft_condensed["Year"] == 2013].copy()
eighth_year = draft_condensed.loc[draft_condensed["Year"] == 2014].copy()
ninth_year = draft_condensed.loc[draft_condensed["Year"] == 2015].copy()
tenth_year = draft_condensed.loc[draft_condensed["Year"] == 2016].copy()
```

```python
# create variables that include list of players from each draft

draft_picks_2007 = first_year["Player"].tolist()
draft_picks_2008 = second_year["Player"].tolist()
draft_picks_2009 = third_year["Player"].tolist()
draft_picks_2010 = fourth_year["Player"].tolist()
draft_picks_2011 = fifth_year["Player"].tolist()
draft_picks_2012 = sixth_year["Player"].tolist()
draft_picks_2013 = seventh_year["Player"].tolist()
draft_picks_2014 = eighth_year["Player"].tolist()
draft_picks_2015 = ninth_year["Player"].tolist()
draft_picks_2016 = tenth_year["Player"].tolist()
```

```python
# import yearly stats 

yearlystats_2008 = pd.read_csv('Data/stats_2008.csv')
yearlystats_2009 = pd.read_csv('Data/stats_2009.csv')
yearlystats_2010 = pd.read_csv('Data/stats_2010.csv')
yearlystats_2011 = pd.read_csv('Data/stats_2011.csv')
yearlystats_2012 = pd.read_csv('Data/stats_2012.csv')
yearlystats_2013 = pd.read_csv('Data/stats_2013.csv')
yearlystats_2014 = pd.read_csv('Data/stats_2014.csv')
yearlystats_2015 = pd.read_csv('Data/stats_2015.csv')
yearlystats_2016 = pd.read_csv('Data/stats_2016.csv')
yearlystats_2017 = pd.read_csv('Data/stats_2017.csv')
yearlystats_2018 = pd.read_csv('Data/stats_2018.csv')
yearlystats_2019 = pd.read_csv('Data/stats_2019.csv')
```

```python
# set up for loop to gather data and create dataframes for 3 years of stats per draft
# This cell is for stats of players drafts in 2008

columns = ['name','team', 'games_played', 'points', 'assists','attempted_field_goals','made_field_goals', 
           'attempted_three_point_field_goals', 'made_three_point_field_goals',
           'attempted_free_throws', 'made_free_throws','offensive_rebounds',
                'defensive_rebounds','steals', 'blocks','turnovers']

drafted_stats_2008 = pd.DataFrame(columns=columns)

name = []
team = []
games_played =[]
points = []
assists = []
attempted_field_goals = []
made_field_goals = []
attempted_three_point_field_goals = []
made_three_point_field_goals = []
attempted_free_throws = []
made_free_throws = []
offensive_rebounds = []
defensive_rebounds = []
steals = []
blocks = []
turnovers = []


for row in yearlystats_2009.itertuples():
    if row[8] in draft_picks_2008:
        name.append(row[8])
        team.append(row[16])
        games_played.append(row[19])
        points.append(row[17])
        assists.append(row[20])
        attempted_field_goals.append(row[6])
        made_field_goals.append(row[12])
        attempted_three_point_field_goals.append(row[1])
        made_three_point_field_goals.append(row[5])
        attempted_free_throws.append(row[18])
        made_free_throws.append(row[7])
        offensive_rebounds.append(row[21])
        defensive_rebounds.append(row[22])
        steals.append(row[15])
        blocks.append(row[2])
        turnovers.append(row[10])
    else: print(f'{row[8]} did not play') 
        
for row in yearlystats_2010.itertuples():
    if row[8] in draft_picks_2008:
        name.append(row[8])
        team.append(row[16])
        games_played.append(row[19])
        points.append(row[17])
        assists.append(row[20])
        attempted_field_goals.append(row[6])
        made_field_goals.append(row[12])
        attempted_three_point_field_goals.append(row[1])
        made_three_point_field_goals.append(row[5])
        attempted_free_throws.append(row[18])
        made_free_throws.append(row[7])
        offensive_rebounds.append(row[21])
        defensive_rebounds.append(row[22])
        steals.append(row[15])
        blocks.append(row[2])
        turnovers.append(row[10])
    else: print(f'{row[8]} did not play') 
        
for row in yearlystats_2011.itertuples():
    if row[8] in draft_picks_2008:
        name.append(row[8])
        team.append(row[16])
        games_played.append(row[19])
        points.append(row[17])
        assists.append(row[20])
        attempted_field_goals.append(row[6])
        made_field_goals.append(row[12])
        attempted_three_point_field_goals.append(row[1])
        made_three_point_field_goals.append(row[5])
        attempted_free_throws.append(row[18])
        made_free_throws.append(row[7])
        offensive_rebounds.append(row[21])
        defensive_rebounds.append(row[22])
        steals.append(row[15])
        blocks.append(row[2])
        turnovers.append(row[10])
    else: print(f'{row[8]} did not play') 
```

```python
drafted_stats_2008["name"] = name
drafted_stats_2008["team"] = team
drafted_stats_2008["games_played"] = games_played
drafted_stats_2008["points"] = points
drafted_stats_2008["assists"] = assists
drafted_stats_2008["attempted_field_goals"] = attempted_field_goals
drafted_stats_2008["made_field_goals"] = made_field_goals
drafted_stats_2008["attempted_three_point_field_goals"] = attempted_three_point_field_goals
drafted_stats_2008["made_three_point_field_goals"] = made_three_point_field_goals
drafted_stats_2008["attempted_free_throws"] = attempted_free_throws
drafted_stats_2008["made_free_throws"] = made_free_throws
drafted_stats_2008["offensive_rebounds"] = offensive_rebounds
drafted_stats_2008["defensive_rebounds"] = defensive_rebounds
drafted_stats_2008["steals"] = steals
drafted_stats_2008["blocks"] = blocks
drafted_stats_2008["turnovers"] = turnovers
```

```python
# Calculations
# average points per game
average_points_2008 = drafted_stats_2008["points"]/drafted_stats_2008["games_played"]
drafted_stats_2008["average_points"] = average_points_2008
```

```python
# field goal percentage
fg_percentage_2008 = drafted_stats_2008['attempted_field_goals']/drafted_stats_2008['made_field_goals']
drafted_stats_2008['fg_percentage'] = fg_percentage_2008
```

```python
# Three point field goal percentage
three_point_percentage_2008 = drafted_stats_2008['attempted_three_point_field_goals']/drafted_stats_2008['made_three_point_field_goals']
drafted_stats_2008['three_point_percentage'] = three_point_percentage_2008
```

```python
# free throw
ft_percentage_2008 = drafted_stats_2008['attempted_free_throws']/drafted_stats_2008['made_free_throws']
drafted_stats_2008 = ft_percentage_2008
```

```python
# rebound average 
average_rebounds_2008 = (drafted_stats_2008['offensive_rebounds'] + drafted_stats_2008['defensive_rebounds'])/drafted_stats_2008['games_played']
drafted_stats_2008['average_rebounds'] = average_rebounds_2008
```

```python
# assist average
average_assists_2008 = drafted_stats_2008['assists']/drafted_stats_2008['games_played']
drafted_stats_2008['average_assists'] = average_assists_2008
```

```python
# turnover average 
average_turnovers_2008 = drafted_stats_2008['turnovers']/drafted_stats_2008['games_played']
drafted_stats_2008['turnovers'] = average_turnoevers_2008
```

```python
# block average
average_blocks_2008 = drafted_stats_2008['blocks']/drafted_stats_2008['games_played']
drafted_stats_2008['average_blocks'] = average_blocks_2008
```

```python
# steal average
average_steals_2008 = drafted_stats_2008['steals']/drafted_stats_2008['games_played']
drafted_stats_2008['steals'] = average_steals_2008
```
